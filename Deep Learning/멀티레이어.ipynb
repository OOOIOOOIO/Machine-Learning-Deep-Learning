# wine 데이터셋을 이용한 퍼셉트론 구현

1. wine 데이터셋을 로드하시오.

2. 로드한 데이터 셋을 학습용/검증용(train/test)으로 나누시오

3. 학습용 데이터를 PCA를 이용하여 8차원으로 차원 축소를 진행하시오,.

4. 8차원으로 축소된 데이터로 와인을 분류하는 퍼셉트론 알고리즘을 학습을 통해 구현하시오.

- 신경망의 구조는 학습자가 다양하게 바꿔가며 성능을 비교해보시오

5. 검증용 셋으로 검증을 수행하시오.

```python
# iris
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

iris_data = load_iris()

X = iris_data.data
y = iris_data.target


X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=1/3)
```

```python
print(y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(X_train)

iris_pca_train = pca.transform(X_train) # 얘가 입력 데이터네
iris_pca_test = pca.transform(X_test) # 얘가 입력 데이터네
```

```python
print(iris_pca_test)
print(iris_pca_train)
```

```python
from tensorflow.keras import models, layers
import numpy as np


# 여러 레이어를 쌓겠다.
model = models.Sequential()

# 한 층을 쌓겠다. (unit : 노드의 개수(출력의 개수), input_shape : 입력의 크기, activation : 전달함수)
model.add(layers.Dense(units = 10,  activation='sigmoid',input_shape=(2,)))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 10,  activation='sigmoid'))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 10,  activation='sigmoid'))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 10,  activation='sigmoid'))

# 히든 레이어 -> 출력
model.add(layers.Dense(units = 1,  activation='sigmoid'))


model.compile(optimizer='adam',
             loss="binary_crossentropy",
             metrics = ['acc'])

model.fit(iris_pca_train, y_train, epochs = 1000, batch_size = 2, verbose = 1)
```

```python
# test
Xtest_loss, Xtest_acc = model.evaluate(iris_pca_test,y_test)
print(Xtest_acc)
```

```python
from tensorflow.keras import models, layers
import numpy as np


# 여러 레이어를 쌓겠다.
model = models.Sequential()

# 한 층을 쌓겠다. (unit : 노드의 개수(출력의 개수), input_shape : 입력의 크기, activation : 전달함수)
model.add(layers.Dense(units = 32,  activation='relu',input_shape=(8,)))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 16,  activation='relu'))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 16,  activation='relu'))

# 히든 레이어 -> 히든레이어
model.add(layers.Dense(units = 8,  activation='relu'))

# 히든 레이어 -> 출력
model.add(layers.Dense(units = 1,  activation='softmax'))


model.compile(optimizer='adam',
  #              loss="binary_crossentropy",
             loss="categorical_crossentropy",
             metrics = ['acc'])

model.fit(wine_pca_train, y_train, epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# test
Xtest_loss, Xtest_acc = model.evaluate(wine_pca_test,y_test)
print(Xtest_acc)
```
