```python
# AND 게이트
from tensorflow.keras import models, layers
import numpy as np



# INPUT
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# OUTPUT(0, 1 -> 2개니까 binary가 적합)
#T = np.array([0,0,0,1]) # AND
#T = np.array([0,1,1,1]) # OR
T = np.array([0,1,1,0]) # XOR, 얘는 무조건 0.5만 뜬다 조심해야됨.
```

```python
# Linear

model = models.Sequential()

model.add(layers.Dense(units = 1,  activation='linear'))

model.compile(optimizer='adam',
              loss="mse", 
              metrics = ['acc'])

model.fit(X,T,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# Sigmoid

model = models.Sequential()

model.add(layers.Dense(units = 64,  activation='relu',input_shape=(2,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(units = 32,  activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(units = 1,  activation='sigmoid'))

model.compile(optimizer='adam',
              loss="binary_crossentropy",
              metrics = ['acc'])

model.fit(X,T,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# softmax

model = models.Sequential()

model.add(layers.Dense(units = 1, activation="softmax", input_shape=(2,)))

model.compile(optimizer="adam",
             #loss = "sparse_categorical_crossentropy",
             loss = "categorical_crossentropy",
             metrics = ["acc"])

model.fit(X, T, epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# Test 하는 법

XTest_loss, XTest_acc = model.evaluate(X, T) # 원래 test 데이터가 들어가야 된다.

print(XTest_acc)
```

```python
# accuracy는 반올림하나보다
Ttest = model.predict(X)
                
print(Ttest)
```
