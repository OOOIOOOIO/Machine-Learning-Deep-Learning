# load ex1 using tf.keras


```python
# adaline using keras
from tensorflow.keras import models, layers
import numpy as np

# AND 게이트

# INPUT / OUTPUT DATA
# in
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# out
T = np.array([0,0,0,1])

# 여러 레이어를 쌓겠다.
model = models.Sequential()

# 한 층을 쌓겠다. (unit : 노드의 개수(출력의 개수), input_shape : 입력의 크기, activation : 전달함수)
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(2,)))

# 전달함수를 sigmoid 함수로 돌려보기, 전달함수의 기본 : 스위치(0 ,1 값을 전달할까 말까), 미분이 가능해야한다.
model.add(layers.Dense(units = 1,  activation='sigmoid',input_shape=(2,)))



# training(파라미터 설정) optimizer와 loss는 그때그떄 좋은 애들을 찾아 바꿔준다.
model.compile(optimizer='sgd', # optimizer은 cost를 줄이기 위한 방법(sgd : 경사하강법) 
                              # 웨이트와 바이어스가 얼마나 좋은지, 작을수록 좋다. adam을 쓰는게 제일 좋다
             loss="mse", # loss는 cost를 구하는 척도이다.(cost func은 y = Wx + b일경우 이차함수가 된다)(loss = target, y의 차)
            loss="binary_crossentropy", # 시그모일 때 제일 적합한 아이, 이진분류일 때는 sigmoid - binary_crossentropy가 짝이다!!!
             metrics = ['acc'])


# epochs(한 주기) = 횟수
# loss가 낮을 수록 좋은 것
model.fit(X,T,epochs = 10000, batch_size = 1, verbose = 1)
```

```python
# test, 원래 test 데이터가 들어가야 된다.
Xtest_loss, Xtest_acc = model.evaluate(X,T)
print(Xtest_acc)
```

```python
# accuracy는 반올림하나보다
Ttest = model.predict(X)
                
print(Ttest) # 0 0 0 1
            # 0 0 / 0 1 / 1 0 / 1 1 일때의 확률
```

# IRIS Data set

##### 1) IRIS 데이터셋을 load하여
##### 2) Train/Test 데이터셋으로 분리하고
##### 3) AND gate와 같은 퍼셉트론 신경망 모델을 만들어
##### 4) 학습 후 test 데이터셋으로 성능을 검증하는 코드를 작성하시오

```python
# adaline using keras
from tensorflow.keras import models, layers
import numpy as np
from sklearn.datasets import load_iris

iris_data = load_iris()

X = iris_data.data
y = iris_data.target
```

```python
# 모델
model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(4,)))

# 학습
model.compile(optimizer='sgd',
             loss="mse",
             metrics = ['acc'])

model.fit(X,y,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# test
Xtest_loss,Xtest_acc = model.evaluate(X,y)
print(Xtest_acc)
```

```python
# accuracy는 반올림하나보다
Ttest = model.predict(X)
print(Ttest) 
```
