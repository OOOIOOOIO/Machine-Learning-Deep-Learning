## wine 데이터셋을 이용한 퍼셉트론 구현

1. wine 데이터셋을 로드하시오.

2. 로드한 데이터 셋을 학습용/검증용(train/test)으로 나누시오

3. 학습용 데이터를 PCA를 이용하여 8차원으로 차원 축소를 진행하시오,.

4. 8차원으로 축소된 데이터로 와인을 분류하는 퍼셉트론 알고리즘을 학습을 통해 구현하시오.

5. 검증용 셋으로 검증을 수행하시오.

```python
# wine
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

wine_data = load_wine()

X = wine_data.data
y = wine_data.target


X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=1/3)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=8)

pca.fit(X_train)

wine_pca_train = pca.transform(X_train) # 얘가 입력 데이터네
wine_pca_test = pca.transform(X_test) # 얘가 입력 데이터네
```

```python
print(wine_data.feature_names)
print(wine_data.target_names)
print(wine_data.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X, columns = wine_data.feature_names)
df["target"] = y

# units의 수
print(len(df["target"].unique()))

# input_shape의 수
print(len(wine_data.feature_names))

df
```

```python
from tensorflow.keras import models, layers
import numpy as np


# 여러 레이어를 쌓겠다.
model = models.Sequential()

model.add(layers.Dense(units = 256, activation="relu", input_shape=(8,)))
#model.add(layers.Dropout(0.5)) # 드롭아웃 : 오버피팅 막는법
model.add(layers.Dense(units = 128, activation="relu"))
model.add(layers.Dense(units = 3,  activation='softmax'))



model.compile(optimizer='sgd',
             loss = "sparse_categorical_crossentropy",
             metrics = ['acc'])

model.fit(wine_pca_train, y_train, epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# test
Xtest_loss, Xtest_acc = model.evaluate(wine_pca_test,y_test)
print(Xtest_acc)
```

```python
# 예측
Ttest = model.predict(wine_pca_test)
                
print(Ttest) 
```

```python
# 행 중에 가장 큰 애를 가져온다.
# 이를 y_test와 비교하면 된다!
np.argmax(Ttest, axis = 1)
```
