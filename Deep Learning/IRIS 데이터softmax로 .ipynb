```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.datasets import load_iris
```

```python
# iris
# multi, output : 0, 1, 2

iris_data = load_iris()

X = iris_data.data
y = iris_data.target

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=1/3)
```

```python
print(iris_data.feature_names)
print(iris_data.target_names)
print(iris_data.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X, columns = iris_data.feature_names)
df["target"] = y

# units의 수
print(len(df["target"].unique()))

# input_shape의 수
print(len(iris_data.feature_names))

df
```

```python
# softmax

model = models.Sequential()

model.add(layers.Dense(units = 3, activation="softmax", input_shape=(4,)))
#model.add(layers.Dropout(0.5)) # 드롭아웃 : 오버피팅 막는법
model.compile(optimizer="adam",
             loss = "sparse_categorical_crossentropy",
             metrics = ["acc"])

model.fit(X_train,y_train,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# Test 하는 법

XTest_loss, XTest_acc = model.evaluate(X_test, y_test) # 원래 test 데이터가 들어가야 된다.

print(XTest_acc)
```

```python
Ttest = model.predict(X_test)
                
print(Ttest)
```

```python
# 행 중에 가장 큰 애를 가져온다.
# 이를 y_test와 비교하면 된다!
np.argmax(Ttest, axis = 1)
```
