```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import numpy as np

```

```python
# breast cancer
# binary, output : 0, 1
from sklearn.datasets import load_breast_cancer

data_cancer = load_breast_cancer();
# X = np.round(data_cancer.data)

# y = np.round(data_cancer.target)

X = data_cancer.data
y = data_cancer.target

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=1/3)
```

```python
print(data_cancer.feature_names)
print(data_cancer.target_names)
print(data_cancer.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X, columns = data_cancer.feature_names)
df["target"] = y

# units의 수
print(len(df["target"].unique()))

# input_shape의 수
print(len(data_cancer.feature_names))

df
```

```python
# Sigmoid

model = models.Sequential()

model.add(layers.Dense(units = 256, activation="relu", input_shape=(30,)))
#model.add(layers.Dropout(0.5)) # 드롭아웃 : 오버피팅 막는법
model.add(layers.Dense(units = 128, activation="relu"))

model.add(layers.Dense(units = 64, activation="relu"))

model.add(layers.Dense(units = 32, activation="relu"))

model.add(layers.Dense(units = 1,  activation='sigmoid'))

model.compile(optimizer='adam',
              loss="binary_crossentropy",
              metrics = ['acc'])

model.fit(X_train,y_train,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# Test 하는 법

XTest_loss, XTest_acc = model.evaluate(X_test, y_test) # 원래 test 데이터가 들어가야 된다.

print(XTest_acc)
```

```python
# accuracy는 반올림하나보다
Ttest = model.predict(X_test)
                
Ttest
```
