```python
# wine
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

cancer_data = load_breast_cancer()

X = cancer_data.data
y = cancer_data.target
# X = np.round(cancer_data.data)
# y = np.round(cancer_data.target)

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=1/3)
```

```python
print(cancer_data.feature_names)
print(cancer_data.target_names)
print(cancer_data.DESCR)
```

```python
import pandas as pd
df = pd.DataFrame(X, columns = cancer_data.feature_names)
df["target"] = y

# units의 수
print(len(df["target"].unique()))

# input_shape의 수
print(len(cancer_data.feature_names))

df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)

pca.fit(X_train)

cancer_pca_train = pca.transform(X_train) # Train입력
cancer_pca_test = pca.transform(X_test) # Test입력 
```

```python
# Sigmoid
from tensorflow.keras import models, layers
import numpy as np

model = models.Sequential()

model.add(layers.Dense(units = 256, activation="relu", input_shape=(10,)))
#model.add(layers.Dropout(0.5)) # 드롭아웃 : 오버피팅 막는법
model.add(layers.Dense(units = 128, activation="relu"))

model.add(layers.Dense(units = 64, activation="relu"))

model.add(layers.Dense(units = 32, activation="relu"))

model.add(layers.Dense(units = 1,  activation='sigmoid'))

model.compile(optimizer='adam',
              loss="binary_crossentropy",
              metrics = ['acc'])

model.fit(cancer_pca_train,y_train,epochs = 100, batch_size = 1, verbose = 1)
```

```python
# Test 하는 법

XTest_loss, XTest_acc = model.evaluate(cancer_pca_test, y_test) # 원래 test 데이터가 들어가야 된다.

print(XTest_acc)
```

```python
# accuracy는 반올림하나보다
Ttest = model.predict(cancer_pca_test)
                
Ttest
```
