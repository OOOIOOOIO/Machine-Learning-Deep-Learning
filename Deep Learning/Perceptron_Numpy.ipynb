cost, loss(error) : 지금 나와야할 값 과 네트워크의 값(의 차이
경사하강법 : cost가 0이 되는 w 찾는 것

```python
# adaline using numpy
import numpy as np

alpha = 0.1 # learning rate? alpha가 크면 변하는 폭이 크다. alpha가 크면 수렴하지 못하고(발산) NAN이 뜨는 경우도 있으며,
            # 반대로 작을 경우 수렴하는 속도가 너무 느리다.
training_endnum = 10000
W = np.random.random((1,3)) # 가중치[k, k, k]
dataset_num = 4

X = np.array([[0, 0, 1], # 입력(입력, 입력, 바이어스) (4 X 3), 바이어스는 1로 두어야 한다. w
              [0, 1, 1],
              [1, 0, 1],
              [1, 1, 1]])

T = np.array([0, 0, 0, 1]) # 결과 (1 X 4)

# 에폭은 학습의 횟수, 배치사이즈(가중치를 한번 업데이트할 때 사용되는 샘플의 수(집단)를 에폭만큼 반복하여 업데이트한다.
# np.transpose : 전치행렬
# np.dot : 넘파이끼리 곱할 때
for epochs in range(0,training_endnum) :
    print("----- %d th epoch -----" % (epochs+1))
    for j in range(0,dataset_num) :
        
        ## implement delta rule : w <- w + alpha * e * x 
        y = np.dot(W, np.transpose(X[j,:])) # W*X (1 X 3) * (1 X 3) 이니까 X를 전치시켜서 (3 X 1)로 만든다.
        e = T[j] - y # 에러(타겟값 - 구한 결과값)
        dW = alpha * e * np.transpose(X[j,:]) # 새로 갱신된 가중치 
        W = W + dW # 가중치는 기본 가중치 + 갱신된 가중치


for i in range(0,dataset_num) :
    y = np.dot(W,np.transpose(X[i,:]))
    print(y)
```
